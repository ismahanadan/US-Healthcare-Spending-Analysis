```python
import pandas as pd
import matplotlib
#Create reference to CSV file
csvpath = "/Users/laurelwilliamson/Desktop/GITHUB/IHME_DEX_1996_2016_Y2020M03D18.CSV"
#Import the CSV into a pandas DataFrame
health_df = pd.read_csv(csvpath)
```

```python
health_df.head()
```

```python
mental_df = health_df.loc[health_df['cause_name'] == "Mental and behavioral disorders"]
```

```python
mental_df.head()
```

```python
mental_gender = mental_df.loc[:, ["year_id", "age_group_name", "sex", "cause_name", "mean_pub", "mean_all", "mean_pri", "mean_oop"]]
```

```python
mental_gender.head()
```

```python
female_mental = mental_gender.loc[mental_gender['sex'] == "Female"]
female_mental2 =  female_mental.loc[:, ["year_id", "mean_all"]]
females1 = female_mental2.plot.bar(stacked=True)
females1
```

```python
male_mental = mental_gender.loc[mental_gender['sex'] == "Male"]
male_mental2 = male_mental.loc[:, ["year_id", "mean_all"]]
males1 = male_mental2.plot.bar(stacked=True)
males1
```
